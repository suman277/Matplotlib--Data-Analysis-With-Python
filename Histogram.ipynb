```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import random
```

```python
py_student=np.random.randint(18,45,(100))
```

```python
ml_studenst=np.random.randint(15,400,(100))
```

```python
print(py_student)
print(ml_studenst)
```

```python
style.use('ggplot')# this method is used to style the graph as well as to give the background grid patterns
bars=[15,20,25,30,35,40,45]# it is a sereis of values in which the graph is plotted in between the values
plt.figure(figsize=(16,9))#it is used to increase the size of the graph
plt.hist(py_student,bars,rwidth=0.5,histtype='barstacked',color='b',label='Python Students',stacked=True)
bars=[0,50,100,150,200,250,300,350,400]
plt.hist(ml_studenst,bars,rwidth=0.5,histtype='barstacked',orientation='horizontal',color='g',label=' ML Students')
#rwidth is used to change the width of the bars
# histtype is used to change the histogram types
#orienation is used to change the oreination of the graph
#label is used to give a labal or name inside the graph
plt.xlabel('Age')# it is used to give a name to the x axis
plt.ylabel('No')# it used to give a name to the y axis
plt.legend()# it is used to give a label inside the graph

plt.show()
```
