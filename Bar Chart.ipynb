# Bar Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
class_section=['DS','AIML','IOT','CYBER SECURITY','WEB DEVELOPMENT']
sec_ds=[12,34,65,34,87]
sec_ai=[23,45,67,34,79]
sec_iot=[23,56,87,34,76]
sec_cs=[45,67,34,32,89]
sec_wb=[12,45,67,34,87]
```

```python
plt.figure(figsize=(16,9))
plt.bar(class_section, sec_ds)
plt.show()
```

```python
#to draw the graph horizontally
plt.figure(figsize=(16,9))
plt.barh(class_section, sec_ds)
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.bar(class_section, sec_ds,width=0.3) # to change the width of the bar in the graph 
plt.show()


```

```python
plt.figure(figsize=(16,9))
plt.bar(class_section, sec_ds,width=0.3, align='edge') #(align is used to place the graph)
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.bar(class_section, sec_ds,width=0.3, align='edge', color='g') #color is used to change the color of the bars
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.bar(class_section, sec_ds,width=0.3, align='edge', color='g', edgecolor='m') #edge color is used to change the color of the edge
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.bar(class_section, sec_ds,width=0.3, align='edge', color='g', edgecolor='m', linewidth=5)# line width is used to change the line width of the graph
plt.show
```

```python
plt.figure(figsize=(16,9))
plt.bar(class_section, sec_ds,width=0.3, align='edge', color='g', edgecolor='m', linewidth=5, alpha=0.5)# alpha is used to control the opacity of the graph
plt.show
```

```python
plt.figure(figsize=(16,9))
plt.bar(class_section, sec_ds,width=0.3, align='edge', color='g', edgecolor='m', linewidth=5, alpha=0.5, linestyle='dotted')#line style is used to change the style of the line
plt.show
```

```python
plt.figure(figsize=(16,9))
plt.bar(class_section, sec_ds,width=0.3, align='edge', color='g', edgecolor='m', linewidth=5, alpha=0.5, linestyle='dotted', label='DataSet')#it is used to give a  name inside the dataset
plt.show
plt.legend(loc=2, fontsize=36)
```

```python
plt.figure(figsize=(16,9))
plt.bar(class_section, sec_ds,width=0.3, align='edge', color='g', edgecolor='m', linewidth=5, alpha=0.5, linestyle='dotted', label='DataSet', visible=False)#it makes the graph unvisibl to the user with the help of visible
plt.show
plt.legend(loc=2, fontsize=36)
```

```python
plt.figure(figsize=(16,9))
style.use('ggplot')
plt.bar(class_section, sec_ds,width=0.3, align='edge', color='g', edgecolor='m', linewidth=5, alpha=0.5, linestyle='dotted', label='DataSet', visible=True)
plt.show
plt.legend(loc=2, fontsize=36)
```

```python
plt.figure(figsize=(16,9))
style.use('ggplot')
plt.bar(class_section, sec_ds,width=0.3, align='edge', color='g', edgecolor='m', linewidth=5, alpha=0.5, linestyle='dotted', label='DataSet', visible=True)
plt.show
plt.legend(loc=2, fontsize=36)
plt.xlabel('Sections', fontsize=20)# it is used to give a name in the x label
plt.ylabel('Students',fontsize=20)# it is used to give a name in the y label
plt.title('Sections Vs Students', fontsize=25)# it is used to give a title to the graph
```
