```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
data=pd.read_csv("C:\\Users\\Suman\\Desktop\\DATA SCIENCE COURSES\\datasets\\interploate.csv")
```

```python
data
```

```python
x=data['TIME']
y=data['STUDY HOURS']
z=data['REVISION']
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,s=200 )# s is used to change the marker size
plt.xlabel('TIME', fontsize=15)
plt.ylabel('STUDY HOURS',fontsize=15)
plt.title('TIME VS STUDY HOURS',fontsize=20)
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,s=200, c='r' )# c is used tok change the colour of the scatter points
plt.xlabel('TIME', fontsize=15)
plt.ylabel('STUDY HOURS',fontsize=15)
plt.title('TIME VS STUDY HOURS',fontsize=20)
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,marker='<' )# marker method is used to change the marker style
plt.xlabel('TIME', fontsize=15)
plt.ylabel('STUDY HOURS',fontsize=15)
plt.title('TIME VS STUDY HOURS',fontsize=20)
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,alpha=0.4)#alpha is used to change the opacity of the scatter marks
plt.xlabel('TIME', fontsize=15)
plt.ylabel('STUDY HOURS',fontsize=15)
plt.title('TIME VS STUDY HOURS',fontsize=20)
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,linewidths=15)#it is used to change the width of the scatter plot
plt.xlabel('TIME', fontsize=15)
plt.ylabel('STUDY HOURS',fontsize=15)
plt.title('TIME VS STUDY HOURS',fontsize=20)
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,edgecolors='g')
plt.xlabel('TIME', fontsize=15)
plt.ylabel('STUDY HOURS',fontsize=15)
plt.title('TIME VS STUDY HOURS',fontsize=20)
```

```python
#to plot two scatter plots in a single graph
plt.figure(figsize=(16,9))
plt.scatter(x,y,linewidths=15)#it is used to change the width of the scatter plot
plt.scatter(x,z,linewidths=20, marker='<', c='r')
plt.xlabel('TIME', fontsize=15)
plt.ylabel('STUDY HOURS VS REVISION',fontsize=15)
plt.title('TIME VS STUDY HOURS',fontsize=20)
```
